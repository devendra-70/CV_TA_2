```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('road_sign.jpg', cv2.IMREAD_GRAYSCALE)

plt.imshow(img, cmap='gray')
plt.title('Original Image')
plt.axis('off')
plt.show()
```

```python
# Convert to float32 (required by Harris detector)
gray = np.float32(img)

# Apply Harris corner detector
dst = cv2.cornerHarris(gray, blockSize=2, ksize=3, k=0.04)

# Result is dilated for better visibility
dst = cv2.dilate(dst, None)

# Threshold to mark the corners in red
img_corners = cv2.cvtColor(img, cv2.COLOR_GRAY2BGR)
img_corners[dst > 0.01 * dst.max()] = [0, 0, 255]

plt.imshow(cv2.cvtColor(img_corners, cv2.COLOR_BGR2RGB))
plt.title('Corners Detected')
plt.axis('off')
plt.show()
```
