```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Load grayscale image
img = cv2.imread('road_sign.jpg', cv2.IMREAD_GRAYSCALE)

plt.imshow(img, cmap='gray')
plt.title('Original Image')
plt.axis('off')
plt.show()
```

```python
# Copy of the original for drawing
img_corners = cv2.cvtColor(img, cv2.COLOR_GRAY2BGR)

# Apply Shi-Tomasi detector
corners = cv2.goodFeaturesToTrack(img, maxCorners=100, qualityLevel=0.01, minDistance=10)
corners = np.int0(corners)

# Draw the corners as circles
for corner in corners:
    x, y = corner.ravel()
    cv2.circle(img_corners, (x, y), 4, (0, 255, 0), -1)

plt.imshow(cv2.cvtColor(img_corners, cv2.COLOR_BGR2RGB))
plt.title('Shi-Tomasi Corners')
plt.axis('off')
plt.show()
```
